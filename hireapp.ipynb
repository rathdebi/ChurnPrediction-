```python
import streamlit as st
import pickle
import numpy as np

model = pickle.load(open('model.pickle','rb'))


def hireright_predict(satisfaction_level,last_evaluation,number_project,average_montly_hours,
                      time_spend_company,Work_accident,promotion_last_5years,sales_IT,sales_RandD,
                      sales_accounting,sales_hr,sales_management,sales_marketing,sales_product_mng,
                      sales_sales,sales_support,sales_technical,salary_high,salary_low,salary_medium):
    
    data = np.array([[satisfaction_level,last_evaluation,number_project,average_montly_hours,
                      time_spend_company,Work_accident,promotion_last_5years,sales_IT,sales_RandD,
                      sales_accounting,sales_hr,sales_management,sales_marketing,sales_product_mng,
                      sales_sales,sales_support,sales_technical,salary_high,salary_low,salary_medium]]).astype(np.float64)
    
    pred = model.predict_proba(data)
    prediction = '{0:.{1}f}'.format(pred[0][0],2)
    return float(prediction)

def main():
    
    st.title("Hire Right Web Application Xebia ML WebApp")
    html_temp = """
    <div style="background-color:#025246 ;padding:10px">
    <h2 style="color:white;text-align:center;">Hire Right Classifier Prediction</h2>
    </div>
    """
    st.markdown(html_temp, unsafe_allow_html=True)

    satisfaction_level = st.text_input("Satisfaction_Level","Type Here")
    last_evaluation = st.text_input("Last_Evaluation","Type Here")
    number_project = st.text_input("Number_Project","Type Here")
    average_montly_hours = st.text_input("Average_Monthly_Hours","Type Here")
    time_spend_company = st.text_input("Time_Spend_Company","Type Here")
    Work_accident = st.text_input("Work_Accident","Type Here")
    promotion_last_5years = st.text_input("Promotion_Last_5Years","Type Here")
    sales_IT = st.text_input("Sales_IT","Type Here")
    sales_RandD = st.text_input("Sales_RandD","Type Here")
    sales_accounting = st.text_input("Sales_Accounting","Type Here")
    sales_hr = st.text_input("Sales_HR","Type Here")
    sales_management = st.text_input("Sales_Management","Type Here")
    sales_marketing = st.text_input("Sales_Marketing","Type Here")
    sales_product_mng = st.text_input("Sales_product_Mng","Type Here")
    sales_sales = st.text_input("Sales_Sales","Type Here")
    sales_support = st.text_input("Sales_Support","Type Here")
    sales_technical = st.text_input("Sales_Technical","Type Here")
    salary_low = st.text_input("Salary_Low","Type Here")
    salary_medium = st.text_input("Salary_Medium","Type Here")
    salary_high = st.text_input("Salary_High","Type Here")
    
    
    
    hireright_html="""  
      <div style="background-color:#F4D03F;padding:10px >
       <h2 style="color:white;text-align:center;"> Employee hire is right</h2>
       </div>
    """
    hirewrong_html="""  
      <div style="background-color:#F08080;padding:10px >
       <h2 style="color:black ;text-align:center;">Employee hire is at risk</h2>
       </div>
    """

    if st.button("Predict"):
        
        output = hireright_predict(satisfaction_level,last_evaluation,number_project,average_montly_hours,
                      time_spend_company,Work_accident,promotion_last_5years,sales_IT,sales_RandD,
                      sales_accounting,sales_hr,sales_management,sales_marketing,sales_product_mng,
                      sales_sales,sales_support,sales_technical,salary_high,salary_low,salary_medium)
        st.success('The probability of employee churn near future {}'.format(output))

        if output > 0.75:
            st.markdown(hirewrong_html,unsafe_allow_html=True)
        else:
            st.markdown(hireright_html,unsafe_allow_html=True)

if __name__=='__main__':
    main()
```
